```python
import requests
from bs4 import BeautifulSoup

# URL target untuk halaman Daily Sale
url = 'https://thisisapril.com/product-category/daily-sale/'

# Mengirimkan request ke halaman
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
response = requests.get(url, headers=headers)

# Memastikan halaman berhasil diambil
if response.status_code == 200:
    print("Sukses mengambil halaman!")
else:
    print(f"Terjadi kesalahan: {response.status_code}")

# Parsing konten halaman menggunakan BeautifulSoup
bs = BeautifulSoup(response.content, 'html.parser')

# Mengambil semua informasi nama produk, harga, diskon, dan link dari halaman
produk_list = bs.find_all('li', class_='product')

# Loop untuk mengambil data produk
for produk in produk_list:
    # Nama produk
    nama_produk = produk.find('h2', class_='woocommerce-loop-product__title')
    nama_produk = nama_produk.get_text(strip=True) if nama_produk else "Nama tidak ditemukan"

    # Harga diskon produk
    harga_diskon = produk.find('ins')
    if harga_diskon:
        harga_diskon = harga_diskon.find('span', class_='amount').get_text(strip=True)
    else:
        harga_diskon = "Tidak ada harga diskon"

    # Harga asli produk
    harga_asli = produk.find('del')
    if harga_asli:
        harga_asli = harga_asli.find('span', class_='amount').get_text(strip=True)
    else:
        harga_asli = "Harga asli tidak ditemukan"

    # Link produk
    link_produk = produk.find('a', class_='woocommerce-LoopProduct-link')['href'] if produk.find('a', class_='woocommerce-LoopProduct-link') else "Link tidak ditemukan"

    # Menampilkan hasil
    print(f"Nama Produk: {nama_produk}")
    print(f"Harga Asli: {harga_asli}")
    print(f"Harga Diskon: {harga_diskon}")
    print(f"Link: {link_produk}")
    print("-" * 40)
```

```python
import requests
from bs4 import BeautifulSoup

# URL target
root = 'https://thisisapril.com/product-category/daily-sale/'

# List untuk menyimpan data produk
titles = []
original_prices = []
discount_prices = []
links = []

# Mengirim permintaan ke URL
url = requests.get(root)
soup = BeautifulSoup(url.text, 'html.parser')

# Mengecek apakah permintaan berhasil
if url.status_code == 200:
    # Mengambil semua produk
    products = soup.find_all("li", {"class": "product"})

    for product in products:
        # Mengambil nama produk
        title = product.find("h2", {"class": "woocommerce-loop-product__title"})
        titles.append(title.text.strip() if title else 'Tidak Ada Nama Produk')

        # Mengambil harga asli
        original_price = product.find("del")
        original_prices.append(original_price.text.strip() if original_price else 'Tidak Ada Harga Asli')

        # Mengambil harga diskon
        discount_price = product.find("ins")
        discount_prices.append(discount_price.text.strip() if discount_price else 'Tidak Ada Harga Diskon')

        # Mengambil link produk
        link_tag = product.find("a", {"class": "woocommerce-LoopProduct-link"})
        links.append(link_tag.get('href') if link_tag else 'Tidak Ada Link')

    # Menampilkan semua data yang diambil
    for i in range(len(titles)):
        print(f"Nama: {titles[i]}, Harga Asli: {original_prices[i]}, Harga Diskon: {discount_prices[i]}, Link: {links[i]}")

else:
    print(f"Gagal mengakses halaman. Status code: {url.status_code}")
```

```python
len(titles), len(original_prices), len(discount_prices), len(links)
```

```python
import pandas as pd

print(f"Titles: {len(titles)}, Original Price: {len(original_prices)}, Discount Price: {len(discount_prices)}, Links: {len(links)}")

min_length = min(len(titles), len(original_prices), len(discount_prices), len(links))

titles = titles[:min_length]
original_prices = original_prices[:min_length]
discount_prices = discount_prices[:min_length]
links = links[:min_length]

df = pd.DataFrame({
    'Nama Product': titles,
    'Harga Product': original_prices,
    'Harga Diskon': discount_prices,
    'Link Product': links
})

df.head(26)
```
